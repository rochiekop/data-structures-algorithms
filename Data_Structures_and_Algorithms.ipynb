## **2. Maximum Value**
___


> Compare arrays[i] with arrays[i + 1] , if arrays[i] > arrays[i + 1] are
exchanged. So continue until the last number, arrays[length - 1] is the
maximum



image.png

```python
def max(data)->list:
  try:
    length = len(data)
    for i in range(0,length-1):
      if data[i] > data[i+1]:
        temp = data[i]
        data[i] = data[i+1]
        data[i+1] = temp
    return data
  except Exception as err:
    print(err)

data = [60,50,95,80,70]
print(f"List Array", max(data))
print(f"Maximum values is ",max(data)[-1])
```

## **3. Bubble Sorting Algorithm**
----


> Compare arrays[j] with arrays[j + 1] , if arrays[j] > arrays[j + 1] are
exchanged.
Remaining elements repeat this process, until sorting is completed.



image.png

```python
def buble_sort(data)->list:
  try:
    length = len(data)
    if length:
        for i in range(0,length-1):
          for j in range(0,length-i-1):
            if data[j] > data[j+1]:
              temp = data[j]
              data[j] = data[j+1]
              data[j+1] = temp
        return data
    else:
      raise ValueError('data must not be empty')
  except (TypeError, ValueError) as e:
    return e

data = [60,50,95,80,70]
print(f"Bubble Sort Array", buble_sort(data))
```

## **4. Minimum Value**
----

> Initial value minIndex=0, j=1 Compare arrays[minIndex] with arrays[j]
if arrays[minIndex] > arrays[j] then minIndex=j, j++ else j++. continue
until the last number, arrays[minIndex] is the Min Value.




image.png

```python
def minimum_value(data)->list:
  try:
    if data:
      minIndex = 0
      for j in range(minIndex,len(data)):
        if data[minIndex] > data[j]:
          minIndex = j
      return data[minIndex]
    else:
      raise ValueError("data must not be empty")
  except (TypeError, ValueError) as e:
    return e


data = [60,50,95,80,70]
print(f"Minimum Value ", minimum_value(data))
```
